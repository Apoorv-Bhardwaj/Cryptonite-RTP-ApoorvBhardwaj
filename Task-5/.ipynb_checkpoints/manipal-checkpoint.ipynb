```python
mpl="/content/manipal.csv"
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf

# Verify GPU for faster training
print("GPU:", "Active" if tf.config.list_physical_devices('GPU') else "Not Active")

# Load and clean names immediately
df = pd.read_csv(mpl,skiprows=3)
rename_dict = {
    'time': 'date',
    'temperature_2m_mean (°C)': 'temp',
    'precipitation_sum (mm)': 'precip',
    'pressure_msl_mean (hPa)': 'pressure',
    'dew_point_2m_mean (°C)': 'dew_point',
    'wind_speed_10m_max (km/h)': 'wind'
}
df = df.rename(columns=rename_dict)
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values('date')
```

```python
# Long-term trend: Whole timeline
plt.figure(figsize=(15, 4))
plt.plot(df['date'], df['temp'], color='orange', linewidth=0.5)
plt.title('Long-Term Temperature Trend (2011 - 2026)')
plt.show()

# Short-term trend: Last 30 days
plt.figure(figsize=(15, 4))
plt.plot(df['date'].tail(30), df['temp'].tail(30), marker='o', color='red')
plt.title('Short Term Trend,Last 30 Days,to check daily cycles')
plt.grid(True, alpha=0.3)
plt.show()
#previous month seasonal phase
#print(df['date'].tail(30), df['temp'].tail(30))
df.loc[df['date'] == pd.to_datetime('2025-12-17'), :]#Univariate, increased radiation, low cloud cover
```

```python
test_days = 366 #good test size
window = 30    # input window

# Train = everything except the last test_days rows
train_df = df.iloc[:-test_days].reset_index(drop=True)

# Features used for inputs
multi_features = ['temp', 'precip', 'pressure', 'dew_point', 'wind']

#minmax
f_min = train_df[multi_features].min().values.astype(float)
f_max = train_df[multi_features].max().values.astype(float)
denom = (f_max - f_min)
denom[denom == 0] = 1.0   # minimal guard

# we need to ssale whole dataset using parameters computed on train_df only
scaled_data = (df[multi_features].values - f_min) / denom

def create_sequences(data, window=30, mode='univariate'):
    X, y = [], []
    for i in range(len(data) - window):
        if mode == 'univariate':
            X.append(data[i : i + window, 0:1])   # only temp as input
        else:
            X.append(data[i : i + window])       # all features as input
        y.append(data[i + window, 0])           # target = temp (index 0)
    return np.array(X), np.array(y)

# Build datasets
X_uni, y_uni = create_sequences(scaled_data, window=window, mode='univariate')
X_multi, y_multi = create_sequences(scaled_data, window=window, mode='multivariate')

# Final split
X_u_train, X_u_test = X_uni[:-test_days], X_uni[-test_days:]
X_m_train, X_m_test = X_multi[:-test_days], X_multi[-test_days:]
y_train, y_test = y_uni[:-test_days], y_uni[-test_days:]
```

```python
#
def train_weather_model(X_tr, y_tr, X_te, m_type='LSTM'):
    model = tf.keras.Sequential([
        tf.keras.Input(shape=(X_tr.shape[1], X_tr.shape[2])),
        tf.keras.layers.LayerNormalization(),
        tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64, return_sequences=True)) if m_type=='LSTM' else
        tf.keras.layers.Bidirectional(tf.keras.layers.GRU(64, return_sequences=True)),
        tf.keras.layers.LSTM(32) if m_type=='LSTM' else tf.keras.layers.GRU(32),
        tf.keras.layers.Dense(1)
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(0.001), loss='huber')
    # verbose=1 so tosee it working
    history = model.fit(X_tr, y_tr, epochs=40, batch_size=32, validation_split=0.1, verbose=1)
    return model.predict(X_te), history

# 1. Baseline (Univariate LSTM) - Uses X_u_train (1 feature)
print("Training Univariate Baseline...")
u_preds, _ = train_weather_model(X_u_train, y_train, X_u_test, 'LSTM')

# 2. Improved (Multivariate LSTM) - Uses X_m_train (5 features)
print("Training Multivariate Improved Model...")
m_preds, m_history = train_weather_model(X_m_train, y_train, X_m_test, 'LSTM')

# 3. Improved (Multivariate GRU)
print("Training Multivariate GRU...")
g_preds, _ = train_weather_model(X_m_train, y_train, X_m_test, 'GRU')
```

```python
# Un-scaling math using f_min/f_max
t_min, t_max = f_min[0], f_max[0]
actual = y_test * (t_max - t_min) + t_min

def get_err(p):
    p_final = p.flatten() * (t_max - t_min) + t_min
    rmse = np.sqrt(np.mean((actual - p_final)**2))
    mae = np.mean(np.abs(actual - p_final))
    return rmse, mae  # error 1

r_u, m_u = get_err(u_preds)
r_m, m_m = get_err(m_preds)
r_g, m_g = get_err(g_preds)

print("\n--- UNIVARIATE VS MULTIVARIATE TABLE ---")
print(f"{'Setup':<15} | {'RMSE':<8} | {'MAE':<8}")
print("-" * 35)
print(f"{'Uni-LSTM':<15} | {r_u:<8.4f} | {m_u:<8.4f}")
print(f"{'Multi-LSTM':<15} | {r_m:<8.4f} | {m_m:<8.4f}")
print(f"{'Multi-GRU':<15} | {r_g:<8.4f} | {m_g:<8.4f}")
```

```python
#new file copy
mpl = "/content/manipal.csv"

df2 = pd.read_csv(mpl, skiprows=3)
def _find_col(substr, cols):
    key = substr.lower().replace('_','').replace(' ','')
    for c in cols:
        if key in c.lower().replace('_','').replace(' ',''):
            return c
    return None

candidates = {
    'time': 'date',
    'temperature_2m_mean': 'temp',
    'precipitation_sum': 'precip',
    'pressure_msl_mean': 'pressure',
    'dew_point_2m_mean': 'dew_point',
    'wind_speed_10m_max': 'wind',
    'cloud_cover_mean (%)': 'cc',
    'shortwave_radiation_sum (MJ/m²)':'radiation',
    'sunshine_duration (s)':'sunshine'

}
rename_map = {}
for k,v in candidates.items():
    found = _find_col(k, df2.columns)
    if found:
        rename_map[found] = v
df2 = df2.rename(columns=rename_map)


df2['date'] = pd.to_datetime(df2['date'], errors='coerce')
df2 = df2.dropna(subset=['date']).sort_values('date').reset_index(drop=True)

# preserve raw fields and ensure numeric types
df2['precip_mm'] = pd.to_numeric(df2['precip'], errors='coerce').fillna(0.0)
df2['temp_raw']  = pd.to_numeric(df2['temp'], errors='coerce')

# Create rain occurrence and log-amount columns
df2['rain_occ'] = (df2['precip_mm'] > 0).astype(int)
# For amount modelling use log1p to compress skew (log(1 + x))
df2['precip_log1p'] = np.log1p(df2['precip_mm']).astype(float)

# features to feed the model (same as earlier but you can add more)
multi_features = [
    'temp_raw',
    'precip_log1p',
    'pressure',
    'dew_point',
    'wind',
    'cc',
    'radiation',
    'sunshine'
]


# forward-fill then mean)
for f in multi_features:
    df2[f] = pd.to_numeric(df2[f], errors='coerce')

# 2) Explicit date-based test split
test_start = pd.to_datetime('2025-01-04')
test_end   = pd.to_datetime('2026-01-04')

# 3) training-only min/max
window = 30
# make feature matrix (unscaled yet)
F = df2[multi_features].values.astype(float)

# targets
target_dates = df2['date'].iloc[window:].reset_index(drop=True)
N_targets = len(target_dates)

# target date strictly before test_start
train_mask = target_dates < test_start
test_mask  = (target_dates >= test_start) & (target_dates <= test_end)

# compute min/max on training portion of raw features

train_indices_for_minmax = [i for i in range(len(F) - window) if target_dates.iloc[i] < test_start]
train_feature_stack = np.stack([F[i : i + window] for i in range(len(F) - window) if target_dates.iloc[i] < test_start], axis=0)

train_flat = train_feature_stack.reshape(-1, train_feature_stack.shape[-1])
f_min = train_flat.min(axis=0)
f_max = train_flat.max(axis=0)
denom = (f_max - f_min)
denom[denom == 0] = 1.0   # guard: avoid zero-division if a feature is constant in train.

# apply min-max scaling (0..1) using train stats
F_scaled = (F - f_min) / denom
F_scaled = np.clip(F_scaled, 0.0, 1.0)  # keep values in [0,1]

# Create sequences function that returns X and y for given target array
def create_sequences_from_features(F_scaled, target_array, window=30):
    Xs = []
    ys = []
    for i in range(len(F_scaled) - window):
        Xs.append(F_scaled[i : i + window])
        ys.append(target_array[i + window])  # target aligned to date at end-of-window
    return np.array(Xs), np.array(ys)

# classification occurrence)
y_occ = df2['rain_occ'].values
X_all, y_occ_seq = create_sequences_from_features(F_scaled, y_occ, window=window)

# regression target (log1p)
y_log = df2['precip_log1p'].values
_, y_log_seq = create_sequences_from_features(F_scaled, y_log, window=window)

# split into train/test using masks on the sequence-level target_dates
seq_target_dates = pd.to_datetime(target_dates.values)
train_idx = np.where(seq_target_dates < test_start)[0]
test_idx  = np.where((seq_target_dates >= test_start) & (seq_target_dates <= test_end))[0]

X_train = X_all[train_idx];    y_occ_train = y_occ_seq[train_idx];    y_log_train = y_log_seq[train_idx]
X_test  = X_all[test_idx];     y_occ_test  = y_occ_seq[test_idx];     y_log_test  = y_log_seq[test_idx]
print("Sequences -> train:", X_train.shape, " test:", X_test.shape)

# For regression, train only on positive amounts (days where next-day precip > 0) to model amount distribution
pos_mask_train = y_log_train > 0.0
X_reg_train = X_train[pos_mask_train]
y_reg_train = y_log_train[pos_mask_train]
print("Regression training on positive-rain samples:", X_reg_train.shape[0])

def weighted_mse(y_true, y_pred):
    # y_true is log1p(mm)
    # convert back to mm to compute weights
    y_mm = tf.math.expm1(y_true)
    weights = 1.0 + 0.05 * y_mm   # heavier rain → higher penalty
    return tf.reduce_mean(weights * tf.square(y_true - y_pred))


# Build models: classifier (binary) and regressor (log1p)
def build_classifier(input_shape):
    m = Sequential([
        layers.Input(shape=input_shape),
        layers.LayerNormalization(),
        layers.Bidirectional(layers.LSTM(64, return_sequences=True)),
        layers.LSTM(32),
        layers.Dense(1, activation='sigmoid')
    ])
    m.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return m

def build_regressor(input_shape):
    m = Sequential([
        layers.Input(shape=input_shape),
        layers.LayerNormalization(),
        layers.Bidirectional(layers.GRU(64, return_sequences=True)),
        layers.GRU(32),
        layers.Dense(1, activation='linear')
    ])
    m.compile(optimizer='adam', loss='huber', metrics=['mse'])
    return m

input_shape = (X_train.shape[1], X_train.shape[2])
clf = build_classifier(input_shape)
reg = build_regressor(input_shape)

# 6) Fit models (small epochs for demo; increase if you have GPU/time)
es = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True)
print("Training classifier (rain occurrence)...")
history_clf = clf.fit(X_train, y_occ_train, epochs=40, batch_size=64, validation_split=0.1, callbacks=[es], verbose=1)

print("Training regressor (amount on log1p scale) on positive-rain samples...")
if X_reg_train.shape[0] > 32:
    history_reg = reg.fit(X_reg_train, y_reg_train, epochs=60, batch_size=32, validation_split=0.1, callbacks=[es], verbose=1)
else:
    # fallback: if too few positive samples, still train but with larger val-split disabled
    history_reg = reg.fit(X_reg_train, y_reg_train, epochs=30, batch_size=8, validation_split=0.0, verbose=1)

# lets Predict on test split: two-stage
occ_probs = clf.predict(X_test).flatten()
occ_preds = (occ_probs >= 0.5).astype(int)

# For amount predictions, predict for all test windows
reg_preds_log = reg.predict(X_test).flatten()
reg_preds_log = np.maximum(reg_preds_log, 0.0)  # guard negative predictions (log1p can't be negative logically, but model might)
reg_preds_mm = np.expm1(reg_preds_log)  # inverse of log1p

# 
precip_pred_final = (occ_preds * reg_preds_mm)

# true precipitation on test (mm)
precip_true = df2['precip_mm'].values[window:][test_idx]

# 8) Metrics (mm)
def rmse(a,b): return np.sqrt(np.mean((a-b)**2))
def mae(a,b):  return np.mean(np.abs(a-b))

rmse_precip = rmse(precip_true, precip_pred_final)
mae_precip  = mae(precip_true, precip_pred_final)
print("\nPRECIPITATION EVAL (two-stage): RMSE = {:.4f} mm | MAE = {:.4f} mm".format(rmse_precip, mae_precip))
```

```python
# ---- THRESHOLD TUNING FOR RMSE (CRITICAL FIX) ----

thresholds = np.linspace(0.1, 0.6, 26)
best_thr = 0.5
best_rmse = np.inf

# true precipitation on test (mm)
precip_true = df2['precip_mm'].values[window:][test_idx]

for thr in thresholds:
    occ_preds_thr = (occ_probs >= thr).astype(int)
    combined = occ_preds_thr * reg_preds_mm
    rmse_val = np.sqrt(np.mean((precip_true - combined)**2))
    if rmse_val < best_rmse:
        best_rmse = rmse_val
        best_thr = thr

print(f"Best threshold = {best_thr:.2f}")
print(f"RMSE after tuning = {best_rmse:.4f} mm")
```
