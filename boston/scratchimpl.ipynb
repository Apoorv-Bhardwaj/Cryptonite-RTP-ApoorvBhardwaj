```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plot
import sklearn
import time
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("bostonhpd.csv")
df.head()
```

```python
df.info()
```

```python
X = df.drop("PRICE", axis=1).values   # features
y = df["PRICE"].values
```

```python
#normalize
X = (X - X.mean(axis=0)) / X.std(axis=0)
```

```python
X = np.c_[np.ones(X.shape[0]), X]
```

```python
np.random.seed(42)
m = len(y)
indices = np.random.permutation(m)
train_size = int(0.8 * m)
train_idx, test_idx = indices[:train_size], indices[train_size:]

X_train, X_test = X[train_idx], X[test_idx]
y_train, y_test = y[train_idx], y[test_idx]
```

```python
w = np.zeros(X_train.shape[1])  # includes bias weight
lr = 0.01
epochs = 1000
```

```python
for epoch in range(epochs):
    y_pred = np.dot(X_train, w)           # predictions
    error = y_pred - y_train              # residuals
    
    # Compute gradients
    dw = (2/len(y_train)) * np.dot(X_train.T, error)
    
    # Update weights
    w -= lr * dw
    
    # Track loss every 100 epochs
    if epoch % 100 == 0:
        mse = np.mean(error**2)
        print(f"Epoch {epoch}, MSE: {mse:.4f}")
```

```python
y_pred_test = np.dot(X_test, w)
mse_test = np.mean((y_test - y_pred_test)**2)
print("\nFinal Test MeanSqErr:", mse_test)
```
