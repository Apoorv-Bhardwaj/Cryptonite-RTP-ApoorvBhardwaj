```python
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
```

```python
Ks = list(range(2,11))
inertias = []
silhs = []
for k in Ks:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labs = km.fit_predict(X_scaled)
    inertias.append(km.inertia_)
    if len(set(labs))>1:
        silhs.append(silhouette_score(X_scaled, labs))
    else:
        silhs.append(np.nan)
print('Ks are', Ks)
print('inertias are', [round(x,2) for x in inertias])
print('silhouettes are', [round(x,4) if not np.isnan(x) else x for x in silhs])
plt.figure(figsize=(10,4))
plt.subplot(1,2,1); plt.plot(Ks, inertias, '-o'); plt.title('elbow'); plt.xlabel('k')
plt.subplot(1,2,2); plt.plot(Ks, silhs, '-o'); plt.title('silhouette'); plt.xlabel('k')
plt.tight_layout(); plt.show()

best_k = int(Ks[np.nanargmax(silhs)])
print('best_k by silhouette:', best_k)

km_final = KMeans(n_clusters=best_k, random_state=42, n_init=20)
k_labels = km_final.fit_predict(X_scaled)

metrics_k = {}
metrics_k['silhouette'] = float(silhouette_score(X_scaled, k_labels)) if len(set(k_labels))>1 else np.nan
metrics_k['davies_bouldin'] = float(davies_bouldin_score(X_scaled, k_labels))
metrics_k['calinski_harabasz'] = float(calinski_harabasz_score(X_scaled, k_labels))
print('KMeans metrics (rounded):', {k: round(v,4) if isinstance(v,float) else v for k,v in metrics_k.items()})
for metric_name, val in metrics_k.items():
    if isinstance(val, float):            
        val_to_print = round(val, 4)      
    else:
        val_to_print = val
    print(f'  {metric_name}: {val_to_print}')
```

```python
centers_pca = pca.transform(km_final.cluster_centers_)
for lab in sorted(set(k_labels)):
    plt.scatter(X_pca[k_labels==lab,0], X_pca[k_labels==lab,1], s=30, alpha=0.7, label=f'C{lab}')
plt.scatter(centers_pca[:,0], centers_pca[:,1], marker='X', s=120, c='k', label='centers')
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left')
plt.title(f'KMeans k={best_k}'); plt.tight_layout(); plt.show()
```

```python
eps_grid = list(np.linspace(0.2,1.2,11)) + [1.5,2.0,2.5]
min_samples = 5
best_db = None
best_score = -999
best_eps = None
for e in eps_grid:
    db = DBSCAN(eps=e, min_samples=min_samples)
    labs = db.fit_predict(X_scaled)
    n_clusters = len(set(labs)) - (1 if -1 in labs else 0)
    if n_clusters >= 2:
        sc = silhouette_score(X_scaled[labs!=-1], labs[labs!=-1])
    if (not np.isnan(sc)) and sc > best_score:
        best_db = db
        best_score = sc
        best_eps = e
if best_db is None:
    best_db = DBSCAN(eps=0.8, min_samples=min_samples)
    best_db.fit(X_scaled)
db_labels = best_db.labels_
chosen_eps = best_eps if best_eps is not None else 0.8
print('chosen eps:', chosen_eps, 'clusters (excl noise):', len(set(db_labels)) - (1 if -1 in db_labels else 0))
mask = db_labels != -1
if mask.sum()>1 and len(set(db_labels[mask]))>1:metrics_db = {'silhouette': float(silhouette_score(X_scaled[mask], db_labels[mask])),
                                                              'davies_bouldin': float(davies_bouldin_score(X_scaled[mask], db_labels[mask])),
                                                              'calinski_harabasz': float(calinski_harabasz_score(X_scaled[mask], db_labels[mask]))}
else:
    metrics_db = {'silhouette': np.nan, 'davies_bouldin': np.nan, 'calinski_harabasz': np.nan}
print('DBSCAN metrics (rounded):', {k: (round(v,4) if isinstance(v,float) else v) for k,v in metrics_db.items()})
```

```python
for lab in sorted(set(db_labels)):
    labname = 'Noise' if lab==-1 else f'C{lab}'
    plt.scatter(X_pca[db_labels==lab,0], X_pca[db_labels==lab,1], s=30, alpha=0.7, label=labname)
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left'); plt.title(f'DBSCAN eps={chosen_eps}'); 
plt.tight_layout(); 
plt.show()
```

```python
k_values = [2,3,4,5,6,7,8]
metrics_all = {}
for k in k_values:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X_scaled)
    inertia = km.inertia_
    silhouette = silhouette_score(X_scaled, labels)
    metrics_all[k] = {'inertia': inertia, 'silhouette': silhouette}
print("k means Metrics Summary")
for k, m in metrics_all.items():
    print(f"k = {k:<2} , inertia = {m['inertia']} , silhouette = {m['silhouette']}")
best_k = max(metrics_all.items(), key=lambda x: x[1]['silhouette'])[0]
print(f"best k based on silhouette score",best_k)
```
